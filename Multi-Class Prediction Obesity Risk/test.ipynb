```python
import numpy as np
import pandas as pd
from statsmodels.tsa.holtwinters import ExponentialSmoothing
import matplotlib.pyplot as plt

# Example monthly demand data for the previous year
data = [120, 130, 145, 160, 150, 170, 180, 190, 200, 210, 215, 220, 170, 130, 175, 110, 120, 130, 180, 200, 300, 210, 225, 270]

# Create a time series
index = pd.date_range(start='2023-01-01', end='2024-12-01', freq='MS')
ts = pd.Series(data, index)

# Plotting the original data
plt.plot(ts, marker='o', color='blue')
plt.title('Monthly Demand Data')
plt.xlabel('Month')
plt.ylabel('Demand')
plt.grid(True)
plt.show()

# Forecasting using Holt-Winters method
model = ExponentialSmoothing(ts
                             , seasonal='mul'
                             , seasonal_periods=12
                             )
model_fit = model.fit()

# Forecast the next 12 months
forecast = model_fit.forecast(12)

# Plotting the forecast
plt.figure(figsize=(10, 6))
plt.plot(ts, marker='o', color='blue', label='Actual Demand')
plt.plot(forecast, marker='x', color='red', label='Forecasted Demand')
plt.title('Demand Forecast for Next Year')
plt.xlabel('Month')
plt.ylabel('Demand')
plt.legend()
plt.grid(True)
plt.show()
```
